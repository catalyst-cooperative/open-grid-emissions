# Compare 930 data to hourly-eGRID

1) Compute correlation and mean difference between the two 
2) Compare before/after time shift fixes, and before/after physics-based cleaning 

```python
import pandas as pd
import plotly.express as px
```

```python
from gridemissions.load import BaData, KEYS
```

```python
orig = BaData("../data/downloads/eia930/chalendar/EBA_raw.csv")
```

```python
ba = "MISO"
new = pd.read_csv(f"../data/results/power_sector_data/{ba}.csv", parse_dates=["datetime_utc"])
```

```python
toplot = new[new.fuel_category == "total"]
toplot = toplot.merge(orig.df[KEYS["E"]["NG"]%ba], how='left', left_on='datetime_utc', right_index=True)

px.line(toplot, y=[KEYS["E"]["NG"]%ba, "net_generation_mwh"], x="datetime_utc")
```

```python
toplot = new[new.fuel_category == "natural_gas"]
toplot = toplot.merge(orig.df[KEYS["E"]["SRC_NG"]%ba], how='left', left_on='datetime_utc', right_index=True)

px.line(toplot, y=[KEYS["E"]["SRC_NG"]%ba, "net_generation_mwh"], x="datetime_utc")
```
